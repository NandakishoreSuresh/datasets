# Statistics

## Mean median mode

```python
import numpy as np
a=np.array([[1,2,3,4],[2,3,4,2]])
```

```python
a1=np.median(a,axis=1)
```

```python
print("Median=",a1)
```

```python
from scipy import stats
```

```python
a2=stats.mode(a,axis=0)
```

```python
print(a2)
```

```python
print("Variance=",np.var(a))
```

```python
print("Standard=",np.std(a))
```

```python
a=[1,2,3,1,2,4,6]
```

```python
def mean(a):
    avg=0
    count=0
    for i in a:
        avg=avg+i
        count=count+1
    mean=avg/count
    print("Mean=",mean)
```

```python
mean(a)
```

## Central tendencies

```python
def central(a):
    print("Mean=",np.mean(a))
    print("Median=",np.median(a))
    print("Mode=",stats.mode(a))
```

```python
central(a)
A=[2,2,2,2,2]
central(A)
```
